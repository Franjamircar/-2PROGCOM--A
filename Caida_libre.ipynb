#INTRODUCCIÓN

Se conoce como caída libre, el movimiento de un cuerpo bajo la acción exclusiva de un campo gravitatorio. En el experimento analizamos el movimiento como perfecto ya que despreciamos la resistencia aerodinámica que presenta el aire al movimiento del cuerpo, analizando lo que pasaría en el vacío.
En el desarrollo de este experimento debíamos tener claros los conceptos del movimiento de caída libre en un cuerpo. En este caso usamos la fórmula [1]:

y(t) = y0 + v0t + 1/2gt2		[1]

En la figura [1] se pueden observar dos casos diferentes del movimiento de caída libre, la variante de los gráficos es V0 la cual varía dependiendo del contexto.
 
Figura [1] Análisis de movimiento en caída libre.

En cinemática el movimiento rectilíneo uniforme es un movimiento con trayectoria recta y velocidad constante ya que no hay aceleración.

Las ecuaciones empleadas en este movimiento son,

Para la posición del objeto en cualquier instante t es la ecuación [2]:

x(t)=x_o+v*(t-t_o)		[2]

Siendo xo la posición inicial de, v la velocidad, t el tiempo y to el tiempo inicial
      

 
Figura [2]. Movimiento rectilíneo uniforme

En el movimiento uniformemente acelerado también se tiene una trayectoria recta pero la velocidad no es necesariamente constante porque existe una aceleración.

La ecuación de la posición del móvil en el instante en este movimiento es:

x(t)=x_o+v_o/t-to)+a/2 〖(t-to)〗^2

#3.1.2 Procedimiento

Siguiendo las indicaciones del profesor se estableció en la fuente de poder una salida de 5 V en el electromagneto. Cuando el interruptor está en la posición “on” el electromagneto sostiene la bola y una vez el cronómetro está listo se apaga el interruptor y el electromagneto deja caer la bola y el cronómetro que está conectado al electromagneto empieza a medir el tiempo to y luego cuando la bola pasa por el “photogate” el cronómetro se reinicia y empieza a medir el tiempo de la caída.

Moviendo el “photogate” a diferentes alturas se pueden tomar los datos necesarios para medir la gravedad.

Empleando el equipo de la Figura [5] que nos fue asignado, seguimos el procedimiento descrito en el manual de laboratorio.
 
Figura 5. Equipo para medidas de movimiento rectilíneo uniforme y uniformemente acelerado

Para el movimiento rectilíneo uniforme:

1. Se colocó sobre la mitad de la longitud del carril un nivel y se manipularon los tornillos de nivelación hasta que quedó nivelado. 
2. Se midió la longitud total del carril y se dividió en 6 segmentos simétricos. 
3. Se colocó el objeto junto al dispositivo de choque ubicado al comienzo del carril. 
4. Se usó un indicador para las posiciones señalada en cada intervalo de distancia. Se soltó la palanca de impulso. 
5. Se anotaron los datos en la tabla correspondiente espacio S y tiempo t. 
6. Se repitió el registro para el mismo desplazamiento una vez más. 
7. Se realizaron los registros para los demás intervalos de espacio llenando la tabla 1, manteniendo la posición del dispositivo de choque constante.
